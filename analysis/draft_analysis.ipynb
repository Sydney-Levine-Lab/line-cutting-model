```python
from pathlib import Path
from sim_analysis import convert_raw_sim_to_clean

RAW_ROOT = Path("../data/raw-data")      # <-- actual location
READ_ROOT = Path("../data/readable-data/")

# Create base output dir if needed
READ_ROOT.mkdir(parents=True, exist_ok=True)

# Convert all *sim* CSVs under raw-data/** into readable-data/**,
# preserving the subfolder structure.
for raw_file in RAW_ROOT.rglob("*.csv"):
    name = raw_file.name

    # Skip anything that isn't a sim output in the nested-Dict format
    # (e.g. collision files, baselines, random other CSVs)
    if not name.endswith(".csv"):
        continue
    if "collision" in name.lower():
        continue
    if ".pddl_" not in name:
        # not one of the MAPNAME.pddl_noise[_runs].csv files
        continue

    # Compute relative path from RAW_ROOT, e.g. "paper_data/yes_line_10_test.pddl_0.0001.csv"
    rel = raw_file.relative_to(RAW_ROOT)

    # Output directory = same subfolder under READ_ROOT
    out_dir = READ_ROOT / rel.parent

    print(f"Converting {rel} → {out_dir}")
    df_clean, out_path = convert_raw_sim_to_clean(raw_file, out_dir=out_dir)
    # Optional: peek at first few rows for the first file or two
    # display(df_clean.head())
```

```python
from pathlib import Path
import pandas as pd

READ_ROOT = Path("../data/readable-data")
paper_dir = READ_ROOT / "paper_data"

for csv_path in paper_dir.glob("*.csv"):
    df = pd.read_csv(csv_path)
    # Make sure noise is string, strip the trailing ".csv" if present
    df["noise"] = df["noise"].astype(str).str.replace(".csv", "", regex=False)
    df.to_csv(csv_path, index=False)
    print("Fixed noise in:", csv_path.name)
```
