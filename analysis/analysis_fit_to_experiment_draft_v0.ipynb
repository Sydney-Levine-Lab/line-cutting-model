```python
def build_design_for_run_type(
    univ_df,
    out_df,
    xp_df,
    aggregate_across_countries=True,
    country=None,
):
    # normalize xp column names
    xp_df = xp_df.rename(columns={"map": "xp_name"}).copy()

    if not aggregate_across_countries and country is not None:
        xp_df = xp_df[xp_df["country"] == country].copy()
        xp_scenario = xp_df.rename(columns={"rating_mean": "rating_mean_country"})
    else:
        xp_scenario = (
            xp_df.groupby("xp_name", as_index=False)
                 .agg(rating_mean=("rating_mean", "mean"))
        )

    scenario_univ = out_df.merge(univ_df, on="map_name", how="left")

    design = xp_scenario.merge(scenario_univ, on="xp_name", how="inner")
    return design
```

```python
MODEL_SPECS = {
    "univ_only": [
        "univ_aggregate_welfare",
    ],
    "outcome_only": [
        "aggregate_welfare",
        "inequality",
        "cardinal_harm",
        "ordinal_harm_blind",
    ],
    "univ_plus_outcome": [
        "univ_aggregate_welfare",
        "aggregate_welfare",
        "inequality",
        "cardinal_harm",
        "ordinal_harm_blind",
    ],
}

def fit_model(design_df, predictors, dv="rating_mean", name="", run_type=""):
    import statsmodels.api as sm

    X = design_df[predictors]
    X = sm.add_constant(X)
    y = design_df[dv]

    m = sm.OLS(y, X).fit()
    summary_row = {
        "model": name,
        "run_type": run_type,
        "predictors": "+".join(predictors),
        "R2": m.rsquared,
        "adj_R2": m.rsquared_adj,
        "AIC": m.aic,
        "BIC": m.bic,
    }
    return summary_row, m
```

```python
run_types = ["kwon2023", "level-0", "level-1"]
rows = []
fitted_models = {}  # if you want to inspect some later

for rt in run_types:
    univ_rt = get_universalization_summary(rt)
    design_rt = build_design_for_run_type(univ_rt, out, xp)

    for name, preds in MODEL_SPECS.items():
        row, m = fit_model(design_rt, preds, name=name, run_type=rt)
        rows.append(row)
        fitted_models[(rt, name)] = m

import pandas as pd
summary_table = pd.DataFrame(rows)
summary_table
```

```python
import statsmodels.api as sm
import pandas as pd

def fit_ols(design_df, predictors, dv="rating_mean", model_name="", run_type=""):
    """
    Fit OLS: dv ~ predictors, return a small summary row and the model object.
    """
    X = design_df[predictors].copy()
    X = sm.add_constant(X)
    y = design_df[dv]

    m = sm.OLS(y, X).fit()

    summary_row = {
        "model": model_name,
        "run_type": run_type,
        "predictors": "+".join(predictors),
        "R2": m.rsquared,
        "adj_R2": m.rsquared_adj,
        "AIC": m.aic,
        "BIC": m.bic,
    }
    return summary_row, m
```

```python
# Outcome-only baseline candidates
BASELINE_SPECS = {
    "agg_only": [
        "aggregate_welfare",
    ],
    "ineq_only": [
        "inequality",
    ],
    "c_harm_only": [
        "cardinal_harm",
    ],
    "o_harm_only": [
        "ordinal_harm_blind",
    ],
    "gini_only": [
        "gini",
    ],
    "agg_plus_ineq": [
        "aggregate_welfare",
        "inequality",
    ],
    "full_outcome": [
        "aggregate_welfare",
        "inequality",
        "cardinal_harm",
        "ordinal_harm_blind",
    ],
}

# Pick one run_type just to build the design df.
# (Outcome metrics don't depend on run_type, so any run_type works for baseline comparison.)
run_type_for_baseline = "kwon2023"

univ_baseline = get_universalization_summary(run_type_for_baseline)
design_baseline = build_design_for_run_type(univ_baseline, out, xp)

baseline_rows = []
baseline_models = {}

for name, preds in BASELINE_SPECS.items():
    row, m = fit_ols(
        design_df=design_baseline,
        predictors=preds,
        model_name=f"baseline_{name}",
        run_type=run_type_for_baseline,
    )
    baseline_rows.append(row)
    baseline_models[name] = m

baseline_summary = pd.DataFrame(baseline_rows)
baseline_summary
```

```python
import itertools

OUTCOME_PREDICTORS = [
    "aggregate_welfare",
    "inequality",
    "cardinal_harm",
    "ordinal_harm_blind",
]

run_type_for_baseline="level-0"

def run_subset_models(design_df, predictors, model_family, run_type="level-1"):
    """
    Fit all non-empty subsets of `predictors`.

    Returns:
        summary_df: one row per model with R2, adj_R2, AIC, BIC
        models: dict mapping tuple(predictors) -> fitted model
    """
    rows = []
    models = {}

    for k in range(1, len(predictors) + 1):
        for combo in itertools.combinations(predictors, k):
            combo = list(combo)
            name = f"{model_family}_" + "+".join(combo)

            row, m = fit_ols(
                design_df=design_df,
                predictors=combo,
                model_name=name,
                run_type=run_type,
            )
            rows.append(row)
            models[tuple(combo)] = m

    summary_df = pd.DataFrame(rows)
    return summary_df, models

outcome_grid, outcome_models = run_subset_models(
    design_df=design_baseline,
    predictors=OUTCOME_PREDICTORS,
    model_family="outcome",
    run_type=run_type_for_baseline,
)

# Look at the best models by AIC (or BIC)
outcome_grid.sort_values("AIC", ascending=True)

PREDICTORS_WITH_UNIV = [
    "aggregate_welfare",
    "inequality",
    "cardinal_harm",
    "ordinal_harm_blind",
    "univ_aggregate_welfare",  # or univ_inequality, etc.
]

univ_grid, univ_models = run_subset_models(
    design_df=design_baseline,
    predictors=PREDICTORS_WITH_UNIV,
    model_family="outcome_plus_univ",
    run_type=run_type_for_baseline,
)

univ_grid.sort_values("AIC", ascending=True)
```

```python
# Choose which baseline you want to use going forward
BASELINE_NAME = "full_outcome"
BASELINE_PREDICTORS = BASELINE_SPECS[BASELINE_NAME]

# Universalization predictors you want to try.
# Comment out ones you don't have yet.
UNIV_METRICS = [
    "univ_aggregate_welfare",
    "univ_inequality",
    "univ_cardinal_harm",
    "univ_ordinal_harm_blind",
]

run_types = ["kwon2023", "level-0", "level-1"]

rows = []
models = {}

for rt in run_types:
    univ_rt = get_universalization_summary(rt)
    design_rt = build_design_for_run_type(univ_rt, out, xp)

    # 1) Baseline-only model for this run_type (outcomes only)
    row_base, m_base = fit_ols(
        design_df=design_rt,
        predictors=BASELINE_PREDICTORS,
        model_name=f"{BASELINE_NAME}_only",
        run_type=rt,
    )
    rows.append(row_base)
    models[(rt, BASELINE_NAME, "none")] = m_base

    # 2) Baseline + each univ metric, one at a time
    for u in UNIV_METRICS:
        # only use the metric if it's actually present in the dataframe
        if u not in design_rt.columns:
            continue

        preds = BASELINE_PREDICTORS + [u]
        row_u, m_u = fit_ols(
            design_df=design_rt,
            predictors=preds,
            model_name=f"{BASELINE_NAME}_plus_{u}",
            run_type=rt,
        )
        rows.append(row_u)
        models[(rt, BASELINE_NAME, u)] = m_u

comparison_table = pd.DataFrame(rows)
comparison_table
```
